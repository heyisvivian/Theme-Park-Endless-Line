```python
#read the csv file name validation_with_predictions.csv
#and print the number of rows and columns in the file
import pandas as pd
df = pd.read_csv('validation_with_predictions.csv')
df
```

```python
#show me the unique values in the column Attraction
df['Attraction'].unique()
```

```python
#select the rows where Holidays_x = 1
df_weekends = df[df['Weekends_x'] == 1]
df_weekdays = df[df['Weekends_x'] == 0]

#for each different attraction, calculate the average of the column Predictions for weekends and weekdays
weekends = df_weekends.groupby('Attraction')['predictions'].mean()
weekdays = df_weekdays.groupby('Attraction')['predictions'].mean()
```

```python
weekends
```

```python
weekdays
```

```python
#plot the line for those two averages
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(weekends, label='Weekends')
plt.plot(weekdays, label='Weekdays')
plt.legend()
plt.show()
```
